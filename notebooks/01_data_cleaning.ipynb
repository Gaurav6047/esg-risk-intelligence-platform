## Step 9 — Filter Modern Analytical Period

The original dataset contained historical records dating back to 1750.

However, early records contained limited environmental and economic reporting.

To improve analytical reliability and consistency, the dataset was filtered to modern reporting years where ESG indicators are more complete and meaningful.

```python
esg_df = esg_df[
    esg_df['year'] >= 1950
]
```

## Step 10 — Validate Dataset After Year Filtering

The dataset was rechecked after filtering to evaluate improvements in data completeness and analytical quality.

```python
esg_df.shape
```

```python
esg_df.isnull().sum()
```

## Step 11 — Remove Critical Missing Records

Rows missing essential environmental indicators such as CO2 emissions and greenhouse gas metrics were removed because they cannot support meaningful ESG risk analysis.

```python
esg_df = esg_df.dropna(
    subset=['co2', 'total_ghg']
)
```

## Step 12 — Reassess Remaining Missing Values

After removing critical missing records, remaining null values were analyzed to determine feature-specific imputation strategies.

```python
esg_df.isnull().sum()
```

## Step 13 — Handle Missing Economic Indicators

Median imputation was applied to economic variables because it provides stability against extreme outliers commonly found in macroeconomic data.

```python
esg_df['gdp'] = esg_df['gdp'].fillna(
    esg_df['gdp'].median()
)

esg_df['population'] = esg_df['population'].fillna(
    esg_df['population'].median()
)
```

## Step 14 — Interpolate Environmental Metrics

Environmental indicators typically evolve gradually over time.

Country-wise interpolation was applied to preserve yearly environmental trends while minimizing information loss.

```python
esg_df = esg_df.sort_values(
    ['country', 'year']
)

environmental_cols = [
    'methane',
    'energy_per_capita',
    'co2_per_capita'
]

esg_df[environmental_cols] = (
    esg_df.groupby('country')[
        environmental_cols
    ].transform(
        lambda x: x.interpolate()
    )
)
```

## Step 15 — Final Missing Value Validation

A final validation step was performed to identify any remaining missing observations before preparing the dataset for downstream analytics and SQL integration.

```python
esg_df.isnull().sum()
```

## Step 16 — Handle Residual Environmental Missing Values

After interpolation and economic imputation, several environmental indicators still contained missing observations.

Instead of removing all incomplete rows, feature-specific domain logic was applied to preserve analytical coverage and minimize unnecessary data loss.

Environmental emission columns were treated separately because missing values may represent:
- negligible industrial activity
- unavailable reporting
- low fossil-fuel dependency

This approach improves dataset retention while maintaining analytical consistency.

### Emission Column Treatment

Remaining missing values in fossil-fuel emission columns were replaced with zero values.

This decision was made because many countries and historical records reported negligible or unavailable industrial fuel emissions.

```python
list(esg_df.columns)
```

```python
emission_cols = [
    'coal_co2',
    'oil_co2',
    'gas_co2'
]

esg_df[emission_cols] = (
    esg_df[emission_cols]
    .fillna(0)
)
```

### Energy Consumption Imputation

Country-wise forward-fill and backward-fill techniques were applied to energy consumption indicators to preserve temporal continuity across yearly observations.

```python
esg_df['energy_per_capita'] = (
    esg_df.groupby('country')[
        'energy_per_capita'
    ]
    .transform(
        lambda x: x.ffill().bfill()
    )
)
```

## Step 17 — Validate Residual Missing Values

The dataset was revalidated after feature-specific environmental imputations to ensure improved completeness and analytical consistency.

```python
esg_df.isnull().sum()
```

## Step 18 — Remove Residual Incomplete Records

These records were removed to finalize a fully analysis-ready ESG dataset.

A small number of residual missing observations remained after feature-specific imputations.
